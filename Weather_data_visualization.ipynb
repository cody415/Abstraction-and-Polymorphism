```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
weather=pd.read_csv("/content/Weather Dataset - Trial Activity DataSet.csv.csv")
```

```python
weather.head()
```

```python
weather.info()
```

```python
sns.barplot(x="humidity",y="temperature",data=weather)
```

```python
sns.distplot(weather["humidity"])
```

```python
sns.distplot(weather["humidity"],kde=False,rug=True)
```

```python
sns.jointplot(x="humidity",y="temperature",data=weather)
```

```python
sns.jointplot(x="humidity",y="temperature",data=weather,kind="hex")
```

```python
sns.jointplot(x="humidity",y="temperature",data=weather,kind="kde")
```

```python
sns.pairplot(weather[["humidity","temperature","air_pollution_index"]])
```

```python
sns.stripplot(x="humidity",y="temperature",data=weather)
```

```python
sns.stripplot(x="humidity",y="temperature",data=weather,jitter=True)
```

```python
sns.swarmplot(x="humidity",y="temperature",data=weather)
```

```python
sns.boxplot(x="humidity",y="temperature",data=weather,hue="weather_type")
```

```python
sns.countplot(weather["weather_type"])
```

```python
sns.pointplot(x="humidity",y="temperature",data=weather,hue="weather_type")
```
